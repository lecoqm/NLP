```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
```

```python
candidate_info = pd.read_csv('data/archelect_search.csv')
```

```python
text_extracted_1981 = pd.DataFrame(columns=['id', 'text'])
files = os.listdir('data/legislatives_1981')
for file in files:
    with open('data/legislatives_1981/' + file, encoding="utf-8") as f:
        text = f.read()
    text_extracted_1981 = pd.concat([text_extracted_1981, pd.DataFrame([[file.replace('.txt', ''), text]], columns=['id','text'])])
text_extracted_1981 = pd.merge(text_extracted_1981, candidate_info, how='inner')
text_extracted_1981.head()
```
