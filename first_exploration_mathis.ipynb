```python
partis = (
    candidate_info["titulaire-soutien"]
    .dropna()
    .astype(str)
    .str.split(";")
    .explode()
    .str.strip()
)

partis = partis[partis != ""]
freq = partis.value_counts()

with open("partis_frequences.txt", "w", encoding="utf-8") as f:
    for parti, n in freq.items():
        f.write(f"{parti}\t{n}\n")
```

```python
years = [1973, 1978, 1981, 1988, 1993]
text_extracted_all = pd.DataFrame(columns=["id", "text", "year"])

for year in years:
    files = os.listdir(f"data/legislatives_{year}")

    for file in files:
        if not file.endswith(".txt"):
            continue
        if "pdfmaster" in file.lower():
            continue

        with open(f"data/legislatives_{year}/" + file, encoding="utf-8") as f:
            text = f.read()
            text = re.sub(r'(?<=\w)-\s+(?=\w)', '', text)
            text = re.sub(r'Sciences\s+Po\s*/\s*fonds\s+CEVIPO[FV]', '', text, flags=re.IGNORECASE)
            text = re.sub(r'[☐□■▪◼◻]', ' ', text)
            text = re.sub(r'(?im)^\s*(imprimerie|imp\.?|impr\.?).*$', '', text)
            text = re.sub(r'(?im)^\s*ne pas jeter sur la voie publique.*$', '', text)
            text = re.sub(r'(?im)^\s*(édité par|imprimé par|tiré par).*$','', text)

        text_extracted_all = pd.concat([
            text_extracted_all,
            pd.DataFrame([[file.replace(".txt", ""), text, year]], columns=["id", "text", "year"])
        ])

text_extracted_all = pd.merge(text_extracted_all, candidate_info, how="inner", on="id")
```

```python
# Extraire les partis / soutiens
partis = (
    text_extracted_all["titulaire-soutien"]
    .dropna()
    .astype(str)
    .str.split(";")
    .explode()
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)

# Enlever les chaînes vides
partis = partis[partis != ""]

# Compter les fréquences
freq = partis.value_counts()

# Transformer en DataFrame
freq_df = (
    freq.rename_axis("parti")
        .reset_index(name="freq")
        .sort_values("freq", ascending=False)
        .reset_index(drop=True)
)

# Exporter en xlsx
freq_df.to_excel("partis_frequences.xlsx", index=False)

# Vérification
print(freq_df.head())
print(f"{len(freq_df)} partis exportés dans 'partis_frequences.xlsx'")
```
